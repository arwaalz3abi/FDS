## <font color = blue size  = 5> **Task 3: Linear Regression (3 Marks)**

**Task Description:**
- Use the following helper functions in your notebook to compute the regression line
- Then answer the questions

```python
def standard_units(arr):
    return (arr - np.mean(arr)) / np.std(arr)
```

```python
def correlation(t, x, y):
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))
```

```python
def slope(t, x, y):
    r = correlation(t, x, y)
    return r * np.std(t.column(y)) / np.std(t.column(x))
```

```python
def intercept(t, x, y):
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))
```

```python
# Q6. Using the helper functions above, compute and print the slope and intercept of the regression line that predicts petal_width from petal_length. Write the full regression equation.

#Computation of slope and printing it
print(f"The slope of the regression line is {slope(iris, "PetalLengthCm", "PetalWidthCm")}")

#Computation of intercept and printing it
print(f"The intercept of the regression line is {intercept(iris, "PetalLengthCm", "PetalWidthCm")}")

#Full regression equation = y = mx+b or as we took it in class (y=b0+b1*x1)
print(f"The full regression equation is: PetalWidthCm = {slope(iris, "PetalLengthCm", "PetalWidthCm")} * PetalLengthCm + {intercept(iris, "PetalLengthCm", "PetalWidthCm")}")


```

```python
# Q7. Use your regression equation to predict the petal width of a flower with petal_length = 4.5 cm. Show your calculation.

#Prediction of petal width for petal length  = 4.5 cm
print(f"For petal length of 4.5 cm, the predicted width is {slope(iris, "PetalLengthCm", "PetalWidthCm") * 4.5 + intercept(iris, "PetalLengthCm", "PetalWidthCm")} cm")
```

```python
# Q8. Compute the correlation coefficient r between petal_length and petal_width. Is the relationship strong or weak? Positive or negative? Explain in 2-3 sentences.

#Computation of correlation coeeficient r and printing it
print(f"The correlation coefficient r between petal_length and petal_width is {correlation(iris, "PetalLengthCm", "PetalWidthCm")}")

#Based on the r value (0.9627570970509661), the correlation is close to +1 indicating a very strong positive
#relationship between petal length and petal width, such that the petal length increases as the petal width increases.
```

## <font color = blue size  = 5> **Task 4: Residual Analysis (4 Marks)**


- Use the regression line you calculated in the following questions.

```python
# Q9. Compute the predicted petal_width for every flower using your regression equation. Add both the predicted values and the residuals (actual - predicted) as new columns to the table. Display the first 8 rows.

#Computation of predicted petal_width for every flower using regression equation
PredictedPetalWidthCm = slope(iris, "PetalLengthCm", "PetalWidthCm") * iris.column("PetalLengthCm") + intercept(iris, "PetalLengthCm", "PetalWidthCm")

#Addition of both predicted values (PredictedPetalWidthCm) and residuals as new column to the table
iris = iris.with_columns("PredictedPetalWidthCm", PredictedPetalWidthCm, "Residuals", iris.column("PetalWidthCm") - PredictedPetalWidthCm)

#Displaying the first 8 rows
iris.show(8)

```

```python
# Q10. Create a scatter plot of the residuals (y-axis) vs petal_length (x-axis). Add a horizontal red dashed line at y = 0. In 2-3 sentences, interpret what this plot tells you about how well the linear model fits the data.

#Scatter plot of residuals (y-axis) vs petal_length (x-axis)
iris.scatter("PetalLengthCm", "Residuals")

#Addition of a horizontal red dashed line at y = 0
plt.axhline(y=0, color="r", ls="--")

#The linear model does not fit the data very well based on this plot (residuals (y-axis) vs petal_length). That is
#because it displays a pattern and an increasing vertical spread as PetalLengthCm increases. While a very good
#linear model should have points scattered randomly with no pattern around the horizontal line y = 0.

```
